## 2. Data Preparation and Cleaning

Firstly we can begin by preparing a singular combined dataset to make our analysis easier. Then we can begin with cleaning the data. Let us try to see if there exist any null values that represent holes in our dataset. This may require our intervention to correct the same. Secondly, we will focus on addressing inherent issues with the as the dataset's creator has mentioned in their note. At the end we will end with a version of the data on which we can perform the Exploratory Data Analysis.

Let's first combine the dataframes using the pd.concat method, which we will have to perform 5 times, since the function only works on two dataframes at a time.

````python
#combining the datasets alphabetically
temp_comb1_df = [bangaloreraw_df, chennairaw_df]
temp1_df = pd.concat(temp_comb1_df, axis=0, join = 'outer')
temp_comb2_df = [temp1_df, delhiraw_df]
temp2_df = pd.concat(temp_comb2_df, axis=0, join = 'outer')
temp_comb3_df = [temp2_df, hyderabadraw_df]
temp3_df = pd.concat(temp_comb3_df, axis=0, join = 'outer')
temp_comb4_df = [temp3_df, kolkataraw_df]
temp4_df = pd.concat(temp_comb4_df, axis=0, join = 'outer')
temp_comb5_df = [temp4_df, mumbairaw_df]
housingraw_df = pd.concat(temp_comb5_df, axis=0, join = 'outer')
housingraw_df
````

Lets also reset the index.

````python
housingraw_df.reset_index(inplace=True, drop=True)
housingraw_df
````

````python
housingraw_df.shape
````

The dataset we will be using initially has **32963 rows and 41 columns.**

````python
housingraw_df.isnull().values.any()
````

Returning `False`  we can see that every cell is filled with data. This will make it easier to further clean the data. Now, the amenities columns which comprises of the majority of the features of this dataset, let us revisit the dataset's creator's note:

"Since for a set of houses, nothing was mentioned about certain amenities, '9' was used to mark such values, which could indicate the absence of information about the apartment but these values don't ascertain the absence of such a feature in real life."

It is my opinion that the right way to approach this analysis isn't to use the amenities columns' values on face value, rather if they have any value more than 0, this will mean that they have that concerned amenity, we will change that value to true instead, and if it is 0, keep it false.

Therefore the columns which will remain unchanged are:

1. Price
2. Area
3. Location
4. No. of Bedrooms
5. Resale
6. city

````python
#changing the amenities to true and false values
for col in housingraw_df.columns[5:-1]:
    housingraw_df[col] = housingraw_df[col].apply(lambda x: False if x == 0 else True)

housingraw_df
````

Finally, the resale column is unnecessary so we can drop it, and declare the final version of the dataframe as housing_df

````python
housing_df = housingraw_df.drop(columns=['Resale']).copy()
housing_df
````

With this the data is cleaned and prepared, we can now move further.

### 1. Scatterplot

**Area v/s Price - All cities**

````python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=housing_df.Area, y=housing_df.Price, hue = housing_df.city, s=100)
plt.title("Price v/s Area for the metro cities");
````

A scatterplot is a type of data visualization that uses dots to represent the values obtained for two different variables - one plotted along the x-axis and the other plotted along the y-axis. It is particularly useful to investigate the relationship or correlation between the two variables, showcasing trends, concentrations, and outliers. A scatterplot can also include a line of best fit to indicate the trend or correlation between the variables.

We see that Delhi stands out as having expensive outliers and generally more expensive houses at almost all levels of square footage. Strangely enough the houses in Mumbai seem to have much fewer outliers and is concentrated at a step more affordable than Delhi.

````python
plt.figure(figsize=(12, 6))
sns.regplot(x=housing_df.Area, y=housing_df.Price)
plt.title("Linear Regression - Price v/s Area");
````

Regplot, or regression plot, is a scatter plot with a regression line (or curve) which represents the best fit to the data points. This plot is used to visualize the linear relationship between two numeric variables, by plotting the data points and using regression analysis to plot the line that best fits the data.

The general trend wee see is an increase in the price with increase in square footage of the house. This is an obvious and expected consequence.

### 2. Pie Chart
**Availability of Amenities**

We will plot multiple charts to showcase how many of each amenity is available with the houses on survey.

````python
def fixlabels(labels):
    return labels

fig, axes = plt.subplots(nrows=18, ncols=2, figsize=(15, 60))

for col in range(4, len(housing_df.columns)-1):
    row_idx = (col-4)//2
    col_idx = (col-4)%2
    labels = fixlabels(housing_df[housing_df.columns[col]].value_counts().index)
    axes[row_idx][col_idx].pie(housing_df[housing_df.columns[col]].value_counts(), labels=labels, autopct='%1.1f%%', startangle=90)
    axes[row_idx][col_idx].set_title(housing_df.columns[col])

fig.suptitle("Availablity of Amenities throughout New Houses in Metro Cities", fontsize=16)
plt.show();
````

A pie chart is a circular statistical graphic which is divided into slices to illustrate numerical proportion. Each slice of the pie represents a category, and the size of the slice is proportional to the percentage that category represents of the total sum.

All the pie charts generated above tackle the availability of a particular amenity. A general overview suggest that on average, 70% of all surveyed houses have access to all the given amenities, which is a great sign for consumers, showcasing that homebuilders seem to be rather focused on additional benefits accruing as well, when purchasing a house. A deeper dive should be undertaken during the question and answer part.

### Q1. What is the average price of houses in each city?
We can calculate the mean of the 'Price' column for each group in the 'city' column using the groupby method in pandas.

We see that Delhi on average is the most expensive metro to live in, followed closely by Mumbai. The affordable metro to live in (relatively speaking) is Kolkata.

2 rooms, not unexpected, rather obvious to anyone who lives in India Haha! This was not a good question to ask honestly speaking.

### Q3. Are there any amenities that are commonly available/unavailable across all houses?

We can use the value_counts method on each of the amenities columns to find out the frequency of "available" and "unavailable" values.

````python
for col in housing_df.columns[5:-1]:
  print(housing_df[col].value_counts())
````

#### We can do better, lets try to figure the most available and least available amenity.

````python
amenities_columns = []
for col in housing_df.columns[5:-1]:
  amenities_columns.append(col)

amenity_availability = housing_df[amenities_columns].sum()

most_common_amenity = amenity_availability.idxmax()
least_common_amenity = amenity_availability.idxmin()
most_common_amenity_count = amenity_availability.max()
least_common_amenity_count = amenity_availability.min()

print(f"The most commonly available amenity is: {most_common_amenity} with {most_common_amenity_count} occurrences.")
print(f"The least commonly available amenity is: {least_common_amenity} with {least_common_amenity_count} occurrences.")
````

Well that's pretty interesting that wifi is the least commonly available amenity. I guess that's because consumers like to have the choice to decide the type of wifi plan they want to own, on their own. The Lift (or Elevator) being the most common amenity, is a rather obvious observation.

### Q4. Is there a correlation between the area of the house and its price?
We did see a linear relationship occur as a result of our plotting of the regression curve. Let us now use the corr() function to see if there exists a real correlation.

````python
correlation = housing_df['Area'].corr(housing_df['Price'])
print(correlation)
````

The corr() function in pandas is used to compute pairwise correlation of columns, excluding NA/null values. The correlation coefficient is a statistical measure that calculates the strength of the relationship between the relative movements of two variables. The values range between -1.0 and 1.0. A correlation of -1.0 shows a perfect negative correlation, while a correlation of 1.0 shows a perfect positive correlation. A correlation of 0.0 shows no linear relationship between the movement of the two variables.

At 0.3, seems to be a rather week correlation between the two, perhaps since the locality is a more important factor in determining housing prices in India and not necessarily the area of the house itself. Also the area of a house in India, across all metros seems to be static, therefore another possible explanation of the weak correlation.

### Q5. Are larger houses (by area) more likely to have more amenities?

We can find the answer by calculating the average area for houses with each possible number of amenities. If larger houses generally have more amenities, we should see a positive trend.

````python
amenities_columns = []
for col in housing_df.columns[5:-1]:
  amenities_columns.append(col)

housing_df['amenities_count'] = housing_df[amenities_columns].sum(axis=1, numeric_only=True)
correlation1 = housing_df['Area'].corr(housing_df['amenities_count'])
print(correlation1)
````

Very very weak correlation, almost negligible. So no, larger houses (by area) are not likely to have more amenities.

#### What about Price?

````python
amenities_columns = []
for col in housing_df.columns[5:-1]:
  amenities_columns.append(col)

housing_df['amenities_count'] = housing_df[amenities_columns].sum(axis=1, numeric_only=True)
correlation2 = housing_df['Price'].corr(housing_df['amenities_count'])
print(correlation2)
````

Slightly more but still a weak correlation as we can see. I'm guessing most new houses come rather standard, filled with all kinds of amenities, and neither pricr or the square footage seems to have any real effect on the amenities available.
