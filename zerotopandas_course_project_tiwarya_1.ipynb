Briefly Describing the Source of the dataset, the **The IEA, International Energy Agency**, is an autonomous agency established in 1974 within the framework of the Organisation for Economic Co-operation and Development (OECD). It was initially created in response to the 1973 oil crisis to help member countries respond to major disruptions in the supply of oil. Over the years, the IEA has expanded its role to become a leading global energy authority, addressing a broad range of energy-related issues including oil, gas, coal, electricity, renewable sources, and energy efficiency. Alongside, the **The Global Annual EV Outlook Report** by the IEA provides an in-depth analysis of the state of electric mobility worldwide, capturing the latest developments and trends in the electric vehicle (EV) market. This includes statistics on electric car stock and sales, market shares, policies, and the global charging infrastructure. The report also offers projections and scenarios about the future of the EV market and its implications for the environment, energy demand, and the broader transportation landscape.


The dataset is freely available at the organizations official website, alongside the flagship annual report on the Global EV Outlook. #2. Data Preparation and Cleaning

Firstly, we can begin with cleaning the data. Let us try to see if there exist any null values that represent holes in our dataset. This may require our intervention to correct the same. Secondly, we will focus on grouping the data. With the different parameters, categories, and regions, this step will be crucial, as only appropriate grouping will result in our analysis making any sense to an observer. Once grouped, we will create separate dataframes that will house such groupings with appropriate names.

````python
gevoraw_df.isnull().values.any()
````

Returning `False`  we can see that every cell is filled with data. This will make it easier to group the data. Our next step will be to group the data into appropriate dataframes. Firstly, we can begin by separating historical data from the projections made

````python
gevohistoric_df = gevoraw_df[gevoraw_df['category']=='Historical'].copy()
gevohistoric_df
````

Now that we have the historical data, we can perform similar operations on the projections.

````python
gevoproj_steps_df = gevoraw_df[gevoraw_df['category']=='Projection-STEPS'].copy()
gevoproj_aps_df = gevoraw_df[gevoraw_df['category']=='Projection-APS'].copy()
gevoproj_steps_df
gevoproj_aps_df
````

In the next step, we will try to further prepare the data by reducing the regions analysed, we can appropriately do so by taking the regions which are common in all 3 of the new dataframes. The main reason we are doing this is to ensure uniformity in our analysis and prevent this from becoming far too massive, considering we have more than 10,000 rows of data. We will use the `.intersection()` function.


*Note to Observer: As a recommendation, feel free to analyse the regions that have been left out of the analysis due to our selection criteria.*

````python
#since we are doing a sets operation,
#we can convert the region column of each dataframe into a set.

set1 = set(gevohistoric_df['region'])
set2 = set(gevoproj_steps_df['region'])
set3 = set(gevoproj_aps_df['region'])

common_regions = set1.intersection(set2).intersection(set3)
common_regions
````

And there we have it! The 6 Regions which will be used for our analysis. They are as follows:
1. China
2. India
3. USA
4. Europe
5. Rest of the World
6. World

In essence, each region offers invaluable insights shaping the global trajectory of electric mobility. We can continue our cleaning by reducing the dataframes to only include the necessary regions. We should also reset the index, as we forgot to do the same earlier.

````python
#we will use the .isin() function
gevoh_df = gevohistoric_df[gevohistoric_df['region'].isin(common_regions)].copy()
gevoh_df

#similarly for the rest
gevop_steps_df = gevoproj_aps_df[gevoproj_aps_df['region'].isin(common_regions)].copy()
gevop_aps_df = gevoproj_aps_df[gevoproj_aps_df['region'].isin(common_regions)].copy()

#now lets reset the index
gevoh_df.reset_index(inplace=True, drop=True)
gevop_steps_df.reset_index(inplace=True, drop=True)
gevop_aps_df.reset_index(inplace=True, drop=True)
````

````python
gevoh_df
````

````python
gevop_steps_df
````

````python
gevop_aps_df
````

Also this should have been taken care earlier, but we learn as we go, the 'Rest of the world' doesn't make sense being in this dataframe, for this region is made in reference to the countries left out when we take into account all of the individual countries. However, we have already dropped a majority of the regions (countries that is), which renders this region misleading, in the context of the regions we have taken in the final version of our dataframes. Therefore, we will exclude this region as well, and compare the remaining directly to the global numbers.

````python
cond1 = gevoh_df['region'] == 'Rest of the world'
cond2 = gevop_steps_df['region'] == 'Rest of the world'
cond3 = gevop_aps_df['region'] == 'Rest of the world'
gevoh_df = gevoh_df[~cond1]
gevop_steps_df = gevop_steps_df[~cond2]
gevop_aps_df = gevop_aps_df[~cond3]
gevoh_df['region'].unique()
````

Finally we can combine the three dataframes and make one combined dataframe, which we may or may not use, but it helps cover all bases.

````python
temp_comb1_df = [gevoh_df, gevop_steps_df]
tempgc_df = pd.concat(temp_comb1_df, axis=0, join = 'outer')
temp_comb2_df = [tempgc_df, gevop_aps_df]
gevocombined_df = pd.concat(temp_comb2_df, axis=0, join = 'outer')
gevocombined_df
````

Last thing, just reset index.

````python
gevocombined_df.reset_index(inplace=True, drop=True)
gevocombined_df
````

#3. Exploratory Analysis and Visualization

Before we begin any attempt of asking pressing questions about our dataset, we need to broadly understand what is represents, and we do this through an Exploratory Data Analysis. An EDA visually and quantitatively inspects, summarizes, and attempts to understand the structure, relationships, anomalies, and patterns in data. In doing so, we will gain valuable insights, and also ensure that subsequent analyses are valid, meaningful, and interpretable.

**Why Conduct EDA on the Global EV Outlook Dataset?**
1. Informed Decision Making: Policymakers, industry leaders, and researchers can make informed decisions based on historical trends and future projections.

2. Understand Market Dynamics: It aids in understanding which regions are adopting EVs faster and why, helping stakeholders in their strategic planning.

3. Resource Allocation: For businesses, understanding the distribution of EV types and powertrains can guide investments.

4. Environmental Insights: Tracking electricity demand and oil displacement helps to assess the environmental implications of the shift to EVs.

5. Technological Advancement: Observing the growth in EV charging points can hint at technological and infrastructural advancements in different regions.

Firstly, I am gonna use Geopandas which I just learnt about to showcase the regions that we have on the map and highlight and regions differently.


````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

````python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
selected_countries = world.loc[world['name'].isin(['China', 'India', 'United States of America'])]
eu_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark',
    'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy',
    'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal',
    'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden'
]

eu = world.loc[world['name'].isin(eu_countries)]

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))
world.boundary.plot(ax=ax, color='lightgray')
selected_countries.plot(ax=ax, color='green')
eu.plot(ax=ax, color='blue')

# Create custom patches for the legend
patch_world = mpatches.Patch(color='lightgray', label='World')
patch_selected_countries = mpatches.Patch(color='green', label='Countries')
patch_eu = mpatches.Patch(color='blue', label='European Union')

# Plot the legend using the custom patches
ax.legend(handles=[patch_world, patch_selected_countries, patch_eu], loc='upper left')

plt.title('Regions')
plt.show()

````

````python
oil_displace_2022 = gevoh_df[(gevoh_df['year']) & (gevoh_df['parameter'].isin(['Oil displacement Mbd', 'Oil displacement million lge']))]

plt.figure(figsize=(14, 7))
sns.boxplot(data=oil_displace_2022, x='region', y='value', hue='parameter')
plt.title('Distribution of Oil Displacement Across Regions historically')
plt.show()
````

````python
# Filter data for years between 2011 and 2022 and 'Oil displacement' parameters
oil_displace_range = gevoh_df[(gevoh_df['year'].between(2011, 2022)) & (gevoh_df['parameter'].isin(['Oil displacement Mbd', 'Oil displacement million lge']))]

# Create a catplot which creates a grid of boxplots, one for each year
g = sns.catplot(
    data=oil_displace_range,
    x='region',
    y='value',
    hue='parameter',
    col='year',
    col_wrap=4,  # Number of columns you'd like before wrapping to a new row. You can adjust this as needed.
    kind='box',
    height=5,
    aspect=1.2  # Adjusts the width of each individual plot
)

g.fig.suptitle('Distribution of Oil Displacement Across Regions from 2011 to 2022', y=1.02)
plt.show()
````

````python
# Assuming df is your DataFrame
# Filter data for 'Oil displacement Mbd' parameter and the region "China"
oil_displace_mbd = gevoh_df[(gevoh_df['parameter'] == 'Oil displacement Mbd') & (gevoh_df['region'] == 'China')]

plt.figure(figsize=(14, 7))
sns.lineplot(data=oil_displace_mbd, x='year', y='value')
plt.title('Oil Displacement Mbd in China from 2011 to 2022')
plt.xlabel('Year')
plt.ylabel('Oil Displacement Mbd')
plt.show()
````

````python
oil_displace_mbd1 = gevoh_df[(gevoh_df['parameter'] == 'Oil displacement Mbd') & (gevoh_df['region'] == 'World')]

plt.figure(figsize=(14, 7))
sns.lineplot(data=oil_displace_mbd1, x='year', y='value')
plt.title('Oil Displacement Mbd in Wordl from 2011 to 2022')
plt.xlabel('Year')
plt.ylabel('Oil Displacement Mbd')
plt.show()
````

````python
grouped = gevoh_df.groupby(['region', 'parameter'])['value'].agg(['sum', 'mean']).reset_index()
sns.barplot(data=grouped, x='region', y='sum', hue='parameter')
plt.title('Total Value by Region per Parameter')
plt.show()
````
